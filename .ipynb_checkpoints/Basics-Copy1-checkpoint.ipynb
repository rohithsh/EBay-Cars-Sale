### Description

Based on the basic info of the data set, it can be easily identified that the dataset contains list of cars that were on sale on eBay when the data was crawled from the website.

```python
autos.columns
```

```python
autos.rename({'yearOfRegistration':'registration_year',
              'monthOfRegistraion':'registration_month',
              'notRepairedDamage':'unrepaired_damage',
              'dateCreated':'ad_created',
              'dateCreated':'date_created',
              'postalCode':'postal_code',
              'lastSeen':'last_seen',
              'vehicleType':'vehicle_type',
              'dateCrawled':'date_crawled',
              'offerType':'offer_type',
              'nrOfPictures':'nr_of_pictures',
              'powerPS':'power_PS',
              'monthOfRegistration':'month_of_registration',
              'fuelType':'fuel_type'},axis=1,inplace=True)
```

```python
autos.drop(['nr_of_pictures','seller','offer_type'],axis=1)
```

```python
autos['odometer'] = (autos['odometer'].str.replace('km','')
                     .str.replace(',','').astype(int))
```

```python
autos['price'] = (autos['price'].str.replace('$','')
                  .str.replace(",","").astype(int))
```

### Exploratory Analysis on Date Columns

As noted, the data set has a lot of columns with date/year values. We explore these cloumns in the above cells.
The **registration_year** column is a label to identify the year in which the car was initially registered. As noted in the above cell, values such as *1000* and *9999* are present in the data set. These are outliers and will affect our analysis. Cleaning these outliers is a part of the work.

```python
autos = autos[autos['registration_year'].between(1900,2020)]
```

```python
(autos['registration_year'].value_counts(normalize=True)
.sort_values().head(5))
```

### Analysis

The outilers in the **registration_year** column have been removed. Now the column in clean. The most no. of cars in the website were registerd in 2003.

```python
(autos['brand'].value_counts(normalize=True)
.sort_values(ascending=False).head(5))
```

```python
brands_counts = autos['brand'].value_counts(normalize=True)
common_brands = brands_counts[brands_counts > 0.05].index
```

```python
print(common_brands)
```

```python
mean_price ={}

for brand in common_brands:
    mean_price[brand] = autos[autos['brand']==brand]['price'].mean()
```

```python
mean_price
```

### Analysing the Price

We dicovered that the common brands of vehicles up for sale are 'audi', 'bmw', 'ford', 'benz', 'opel' and 'volkswagen'. For these common brands, we analyse the average price of the cars listed in the webpage. It is analysed that 'audi' brand cars have an average price that is more than 3 times the average price of 'opel' brand cars.

```python
mean_mileage= {}

for brand in common_brands:
    mean_mileage[brand] = (autos[autos['brand']==brand]['odometer_km']
                         .mean())
```

```python
mean_mileage
```

```python
mileage_info = pd.Series(mean_mileage).sort_values()
mileage_info
```

```python
price_info = pd.Series(mean_price).sort_values()
price_info
```

```python
common_brand_info = pd.DataFrame(mileage_info, columns=['Mean Mileage'])
```

```python
common_brand_info['Mean Price'] = price_info
```

```python
common_brand_info
```

### Replacing German words with English words

In the following cells, we replace some common words used in a lable that are in German with their equivalent English words.

```python
autos.head(2)
```

```python
autos['unrepaired_damage'].unique()
```

```python
autos['unrepaired_damage'] = (autos['unrepaired_damage'].str
                              .replace('nein','no').str
                              .replace('ja','yes'))
```

```python
autos['unrepaired_damage'].unique()
```

```python
autos['gearbox'].unique()
```

```python
autos['gearbox'] = (autos['gearbox'].str.replace('manuell','manual')
                    .str.replace('automatik','automatic'))
```

```python
autos['gearbox'].unique()
```

### Comparing Damaged vs Undamaged Cars

```python
damaged_cars_price = int(autos[autos['unrepaired_damage'] == 'yes']
                      ['price'].mean())
damaged_cars_price
```

```python
undamaged_cars_price = int(autos[autos['unrepaired_damage'] == 'no']
                      ['price'].mean())
undamaged_cars_price
```

### Analysis

The average selling price of a damaged car is 3 times lesser than that of an undamaged car.
